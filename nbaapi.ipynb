**Find players and player IDs**
https://github.com/swar/nba_api/blob/master/docs/table_of_contents.md

```python
from nba_api.stats.static import players
import pandas as pd
```

**Create a dictionary to get players**

```python
player_dict = players.get_players()
```

**Use a for loop to find players by their "full_name"**

```python
goat = [player for player in player_dict if player['full_name'] == 'LeBron James'][0]
goat_id = goat['id']
```

```python
print(goat_id)
```

**Create a dictionary to get teams**

```python
from nba_api.stats.static import teams
team_dict = teams.get_teams()
```

**Use a for loop to find teams by their "team_name"**

```python
celtics = [team for team in team_dict if team['full_name'] == 'Boston Celtics'][0]
celtics_id = celtics['id']
```

```python
print(celtics_id)
```

**Retreiving game logs for a specific season and player**

```python
from nba_api.stats.endpoints import playergamelog

gamelog_goat = playergamelog.PlayerGameLog(player_id = '2544', season = '2024-25')
print(gamelog_goat)

gamelog_goat_df = gamelog_goat.get_data_frames()[0]

gamelog_goat_df.head()
```

**Retrieving game logs for all seasons for a specific player**

```python
from nba_api.stats.library.parameters import SeasonAll

gamelog_goat_all_seasons = playergamelog.PlayerGameLog(player_id=goat_id, season = SeasonAll.all)
df_goat_games_all_seasons = gamelog_goat_all_seasons.get_data_frames()[0]
df_goat_games_all_seasons.head()
```

**Retrieving all the game logs from a team**

```python
from nba_api.stats.endpoints import leaguegamefinder
celtics_games = leaguegamefinder.LeagueGameFinder(team_id_nullable = celtics_id).get_data_frames()[0]
celtics_games.head()
```
